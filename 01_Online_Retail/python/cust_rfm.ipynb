```python
import pandas as pd
```

```python
cust_rfm=pd.read_csv(r"D:\내 자료실(이동,삭제 절대금지~~!!)\Desktop\data analysis\proj_3 online_retail\python\cust_rfm.csv")
```

```python
cust_rfm["R"] = pd.qcut(
    cust_rfm["recency"],
    q=5,
    labels=[5,4,3,2,1])
```

```python
pd.qcut(
    cust_rfm["frequency"],
    q=5,
    duplicates="drop"
)
```

```python
def f_score(x):
    if x == 1:
        return 1
    elif x <= 3:
        return 2
    elif x <= 6:
        return 3
    elif x <= 15:
        return 4
    else:
        return 5

cust_rfm["F"] = cust_rfm["frequency"].apply(f_score)
```

```python
cust_rfm["M"] = pd.qcut(
    cust_rfm["monetary"],
    q=5,
    labels=[1,2,3,4,5]
)
```

```python
cust_rfm.head()
```

```python
import numpy as np

conditions = [
    (cust_rfm["R"] >= 4) & (cust_rfm["F"] >= 4) & (cust_rfm["M"] >= 4),
    (cust_rfm["R"] >= 4) & ((cust_rfm["F"] >= 4) | (cust_rfm["M"] >= 4)),
    (cust_rfm["R"] <= 2) & (cust_rfm["F"] >= 4)
]

choices = [
    "High Value Customers",
    "Potential High Value Customers",
    "At Risk Customers"
]

cust_rfm["segment"] = np.select(
    conditions,
    choices,
    default="Regular Customers"
)
```
