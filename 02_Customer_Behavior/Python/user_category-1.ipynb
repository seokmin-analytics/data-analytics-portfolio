```python
# event_cnt 이상치 탐색

user_category_df["event_cnt"].value_counts().sort_index(ascending=True).head(500)
```

```python
# 그래프 그리기전 임시 프레임

tmp = (user_category_df.groupby("event_cnt")["purchase_flag"]
      .mean().reset_index())
```

```python
tmp = tmp[tmp["event_cnt"] <= 169]
```

```python
# event_cnt 와 구매의사비율 관계

import matplotlib.pyplot as plt

plt.plot(
    tmp["event_cnt"],
    tmp["purchase_flag"]
)

plt.xlabel("event_cnt")
plt.ylabel("purchase_rate")

plt.show()
```
